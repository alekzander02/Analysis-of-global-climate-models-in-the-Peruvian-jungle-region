```python
import xarray as xr
from datetime import datetime, timedelta
import pandas as pd
```

```python
!pip install netCDF4
```

```python
tmax = xr.open_dataset("/content/drive/MyDrive/Tecnicas Estadisticas/PISCOdtx_v1.1.nc")
tmax
```

```python
# Lista de coordenadas de localidades (latitud, longitud)
localidades = [(-4.003, -73.161), (-6.915, -76.799), (-5.949, -76.085), (-4.519, -73.6), (-5.043, -73.836), (-4.9, -73.65), (-6.593, -76.322), 
 (-6.406, -76.61), (-6.356, -75.779)]

# Rango de años
rango = pd.date_range(start=datetime(1981, 1, 1), end=datetime(2013, 12, 31))

# Crear un DataFrame vacío
df = pd.DataFrame({'Fecha': rango})

# Iterar sobre las coordenadas de localidades y obtener los valores de precipitación para cada una
for latitud, longitud in localidades:
    # Extraer la variable de precipitación para la ubicación específica y rango de años
    variable_tx = tmax['tx'].sel(latitude=latitud, longitude=longitud,time = rango, method="nearest")
    
    # Obtener los valores de precipitación diarios
    datos_diarios = variable_tx.values.flatten()
    
    columna_nombre = f'Tmax_{latitud}_{longitud}'
    
    # Agregar una columna al DataFrame con los valores de precipitación correspondientes
    df[columna_nombre] = datos_diarios

# Imprimir el DataFrame 
df
```

```python
df.columns
```

```python
df = df.rename(columns={'Tmax_-4.003_-73.161': 'ho00000172', "Tmax_-6.915_-76.799": "ho00000211","Tmax_-5.949_-76.085":"ho00000278",
          "Tmax_-4.519_-73.6":"ho00000279", "Tmax_-5.043_-73.836":"ho00000280", "Tmax_-4.9_-73.65": "ho00000281", 
          "Tmax_-6.593_-76.322": "ho00000310", "Tmax_-6.406_-76.61": "ho00000322", "Tmax_-6.356_-75.779": "ho00000386"})
df
```

```python
df =df.round(decimals=1)
df
```

```python
df.to_csv(r"tmax_1981-2013_series.csv", index =False)
```
